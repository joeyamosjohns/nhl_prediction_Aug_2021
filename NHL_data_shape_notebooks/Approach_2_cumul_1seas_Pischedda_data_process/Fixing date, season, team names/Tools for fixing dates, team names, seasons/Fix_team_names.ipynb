```python

def bet_mp_date(date, season):
    d = str(date)
    s = str(season)
    if len(d) == 3:
        d = "0"+d 
    if (900 < date):
        y= s[:4]
    else:
        y = s[4:]
    return int(y+d)
```

```python
#the function appends 0 in front of any day or month with one digit to follow mp format

def to_string(n):
    s = str(n)
    if len(s) ==1:
        return "0"+s
    else:
        return s

#takes in date_time date and returns yyyymmdd eg 20080904

def game_add_date(date_time):
    d = date_time.day
    m = date_time.month
    y = date_time.year
    dd = to_string(d)
    mm = to_string(m)
    yyyy = str(y)
    return(int(yyyy+mm+dd))
    
```
